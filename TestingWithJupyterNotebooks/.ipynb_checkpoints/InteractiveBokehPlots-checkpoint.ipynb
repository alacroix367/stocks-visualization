```python
#Import packages
# Helpful resources:
#    http://biobits.org/bokeh-flask.html
#    https://bokeh.pydata.org/en/latest/docs/user_guide/notebook.html
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
N=4000
```

```python
x = np.random.random(size=N)*100
y = np.random.random(size=N)*100
radii = np.random.random(size=N)*1.5
colors = ["#%02x%02x%02x" % (int(r),int(g),150) for r,g in zip(np.floor(50+2*x), np.floor(30+2*y))]
```

```python
p=figure()
p.circle(x,y,radius=radii,fill_color=colors,fill_alpha=0.6,line_color=None)
show(p)
show(p) # just call show() multiple times to show multiple plots
```

```python
## Interactive plots in jupyter notebook + Bokeh
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
```

```python
x = np.linspace(0,2*np.pi,2000)
y = np.sin(x)
```

```python
p = figure(plot_height=300, plot_width=600, y_range=(-5,5))
r = p.line(x,y,color="#2222aa",line_width=3)
show(p)
```

```python
def update(f, w=1, A=1, phi=0):
    if f == "sin" : func = np.sin
    elif f == "cos" : func = np.cos
    elif f == "tan" : func = np.tan
    r.data_source.data['y'] = A*func(w*x+phi)
    push_notebook()
```

```python
show(p,notebook_handle=True)
```

```python
interact(update,f=["sin","cos","tan"],w=(0,100),A=(1,5),phi=(0,20,0.1));
```
